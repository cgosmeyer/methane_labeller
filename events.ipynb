```python
import glob
import numpy as np
import proplot as pplt
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
def fname2detail(fname):
    _type = fname.split('/')[1]
    _datetime = fname.split('/')[3].split('_')[2]
    return datetime.strptime(_datetime, '%Y%m%d'), _type
```

```python
# Algeria
plt.figure(figsize=(20, 2.5))

s2list = glob.glob('classified/cloudy/5.906_31.659/S2*/S2*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [0.2]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='C0-')

s2list = glob.glob('classified/positive/5.906_31.659/S2*/S2*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [0.2]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='k-')

s2list = glob.glob('classified/negative/5.906_31.659/S2*/S2*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [0.2]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='r-')



s2list = glob.glob('classified/cloudy/5.906_31.659/L7*/L7*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [1]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='C0-', bottom=0.8)

s2list = glob.glob('classified/positive/5.906_31.659/L7*/L7*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [1]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='k-', bottom=0.8)

s2list = glob.glob('classified/negative/5.906_31.659/L7*/L7*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [1]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='r-', bottom=0.8)


s2list = glob.glob('classified/cloudy/5.906_31.659/L8*/L8*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [0.6]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='C0-', bottom=0.4)

s2list = glob.glob('classified/positive/5.906_31.659/L8*/L8*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [0.6]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='k-', bottom=0.4)

s2list = glob.glob('classified/negative/5.906_31.659/L8*/L8*npz')
s2results = [fname2detail(f)[1] for f in s2list]
s2datetime = [fname2detail(f)[0] for f in s2list]
plt.stem(s2datetime, [0.6]*len(s2datetime), markerfmt=' ', basefmt=" ", linefmt='r-', bottom=0.4)

plt.gca().get_yaxis().set_visible(False)
plt.text(-0.05, 0.15, 'Sentinel-2', ha='center', va='center', transform=plt.gca().transAxes, fontsize=20)
plt.text(-0.05, 0.5, 'Landsat 8', ha='center', va='center', transform=plt.gca().transAxes, fontsize=20)
plt.text(-0.05, 0.82, 'Landsat 7', ha='center', va='center', transform=plt.gca().transAxes, fontsize=20)
plt.ylim((-0.1, 1.1))
plt.title('Algeria')
plt.savefig('algeria_detect.png', dpi=300)
```
